```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ÏòàÏ†ú Îç∞Ïù¥ÌÑ∞ ÏÉùÏÑ±
np.random.seed(0)
data = pd.DataFrame({
    "feature_1": np.random.normal(loc=0, scale=1, size=100),
    "feature_2": np.random.normal(loc=5, scale=2, size=100)
})
```

```python
plt.figure(figsize=(6, 4))
plt.hist(data["feature_1"], bins=20, color="skyblue", edgecolor="black")
plt.title("Histogram of Feature 1")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(data=data)
plt.title("Boxplot of Feature 1 & Feature 2")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(x="feature_1", y="feature_2", data=data)
plt.title("Scatter Plot of Feature 1 vs Feature 2")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.grid(True)
plt.show()
```
